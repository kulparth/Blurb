```python
import pandas as pd

## Read input data
df = pd.read_csv('df_input.txt', sep='|')
```

```python
## print what we just read
df
```

```python
# time to shape genre_ids and store the out in df1
df1 = df.genre_ids.str.split(',', expand=True)
df1
```

```python
# about time to shape geners and store the out in df2
df2 = df.genres.str.split(',', expand=True)
df2
```

```python
# put the data back in df and store the out in df3
df3 = pd.concat([df, df1], axis=1)
df3
```

```python
# transpose the gener_ids and store the out in df4
df4 = pd.melt(df3, id_vars=['country', 'genre_ids'], value_vars=[0,1,2])
df4
```

```python
# essentially same step as df3 but for generes and store out in df5
df5 = pd.concat([df, df2], axis=1)
df5
```

```python
# transpose data for generes (same step as df4)
df6 = pd.melt(df5, id_vars=['country', 'genre_ids'], value_vars=[0,1,2])
df6
```

```python
# clean up noise, can also be done in earlier steps
df44 = df4.dropna()
df44
```

```python
df66 = df6.dropna()
df66
```

```python
# time to put together shaped up data
df7 = pd.merge(df44, df66, how='inner', on=['country', 'genre_ids', 'variable'])
df7
```

```python
## final out in d77
df77 = df7[['country', 'value_x', 'value_y']].drop_duplicates()
df77
```
